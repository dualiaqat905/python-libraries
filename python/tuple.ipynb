# *tuple*

### Q:Tuple kya hota hai?

#### Ans:Tuple bhi list jaisa hota hai — lekin change nahi ho sakta.

```python
point = (10, 20)
```

### Access Elements

```python
print(point[0])
```

### Tuple Modify nahi ho sakta

```python
point[0] = 50
```

### Single Element Tuple

```python
t = (5,)
```

### Tuple Packing & Unpacking

```python
data = (1, 2, 3)

a, b, c = data

print(a, b, c)
```

# Tuple Methods

### count()

```python
t = (1, 2, 3, 2, 2, 5)

print(t.count(2))
```

### index()

```python
t = (10, 20, 30, 20)

print(t.index(20))
```

```python
t = (5, 10, 5, 20, 5)

print(t.count(5))   # 3
print(t.index(10))  # 1
print(len(t))       # 5
print(20 in t) 
```
