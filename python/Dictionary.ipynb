# *Dictionary (dict)*

```python
student = {
    "firstname": "Ahmad",
    "lastname": "Ali",
    "age": 20,
    "marks": 85.5,
    "fruits":["Ali",1,2,3,4]
}

print(student["firstname"],student["lastname"],student["age"],student["marks"],student["fruits"][0])
print(student["lastname"])
```

### 🔹 Dictionary Methods

#### 1) keys()
##### Keys return karta hai

```python
print(student.keys())
```

#### 2) values()

##### Values return karta hai

```python
print(student.values())
```

#### 3) items()

##### Key + value pairs deta hai

```python
print(student.items())
```

#### 4) get()

##### Safe way to access value

```python
print(student.get("age"))
```

#### 5) update()
##### Dictionary update karta hai

```python
student.update({"age": 21})
print(student.get("age"))
```

#### 6) pop()

##### Specific key remove karta hai

```python
student.pop("marks")
```

#### 7) clear()
##### Sab remove

```python
student.clear()
```
