# Set Operations (Math jaisi)

#### Union (Combine)

```python
A = {1,2,3}
B = {3,4,5}

print(A.union(B))
```

#### Intersection (Common elements)

```python
print(A.intersection(B))
```

#### Difference

```python
print(A.difference(B))
```
