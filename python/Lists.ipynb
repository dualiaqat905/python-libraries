### Access Elements (Indexing)

```python
nums = [5, 10, 15,"Al", 20]

print(nums[3])   # 5
print(nums[-1])  # 20
```

### Slicing

```python
print(nums[1:3])
```

### Modify List

```python
nums[0] = 99
```

```python
print(nums)
```

# List Methods

### append() — end me add

```python
nums.append(25)
```

```python
print(nums)
```

### insert() — specific position pe

```python
nums.insert(1, 50)
```

```python
print(nums)
```

### remove() — value delete

```python
nums.remove(10)
```

```python
print(nums)
```

### pop() — index se delete

```python
num = [1,2,3,4,5]
x = num.pop()
print(x)
print(num)
```

### len() 

```python
len(nums)
```

### sort()

```python
nums.sort()
```

```python
print(nums)
```
