```python
# 1) Function jo "Hello" print kare
def say_hello():
    print("Hello")

say_hello()
```

```python
# 2) Function jo name input le aur greet kare
name = input("Enter Your Name")
def greet(name):
    print("Hello", name)

greet(name)
```

```python
# 3) Function jo 2 numbers ka sum return kare
def add(a, b):
    return a + b

print(add(2, 3))
```

```python
# 4) Function jo number square kare
def square(n):
    return n * n

print(square(4))
```

```python
# 5) Function jo even ya odd check kare
def even_or_odd(n):
    if n % 2 == 0:
        return "Even"
    else:
        return "Odd"

print(even_or_odd(7))
```

```python
# 6) Function jo 3 numbers me se largest return kare
def largest(a, b, c):
    return max(a, b, c)
# max() Python ka built-in function hai jo multiple values me se sab se bari (largest) value return karta hai.
print(largest(5, 9, 2))
```

```python
# 7) Function jo list ka sum calculate kare
def list_sum(lst):
    return sum(lst)

print(list_sum([1,2,3]))
```

```python
# Function jo list ka sum calculate kare

lst = list(map(int, input("Enter numbers separated by space: ").split()))

def list_sum(lst):
    return sum(lst)

print(list_sum(lst))
```

```python
# 8) Function jo string reverse kare
def reverse_string(s):
    return s[::-2]
# string[start : end : step]
print(reverse_string("khadija"))
```
