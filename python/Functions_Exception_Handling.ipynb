#### Variable Scope

#### Variable kaha accessible hai

#### Local Variable

#### Function ke andar hoti hai
##### Outside use nahi ho sakti ❌

```python
def test():
    x = 20
    print(x)

test()
# print(x)
```

#### Global Variable

#### Function ke bahar define hoti hai

```python
x = 5

def show():
    print(x)

show()
```

#### Global modify karna

```python
x = 10

def change():
    global x
    x = 20

change()
print(x)
```

# Lambda Expression

```python
# Normal Function
x = eval(input("ENter a no. "))
def square(x):
    return x * x
print(square(x))
square(x)
```

```python
square = lambda x: x * x
print(square(4))
```

# map() Function

#### 👉 Function ko har element par apply karta hai

```python
nums = [1, 2, 3, 4]

result = list(map(lambda x: x*2, nums))
print(result)
```

```python
nums = [1, 2, 3, 4]

result = map(lambda x: x * 2, nums)

print(result)
```

```python
nums = [1, 2, 3, 4]

result = map(lambda x: x * 2, nums)

print(list(result))
```

# filter() Function

#### 👉 Condition True ho to element keep hota hai

```python
nums = [1, 2, 3, 4, 5]

result = list(filter(lambda x: x % 2 == 0, nums))
print(result)
```

```python
nums = [1, 2, 3, 4, 5]

result = filter(lambda x: x % 2 == 0, nums)
print(list(result))
```

# Inner / Nested Functions

#### 👉 Function ke andar function

```python
def outer():
    
    def inner():
        print("Inner function")
        
    inner()

outer()
```

# Exception Handling

```python
try:
    x = 10 / 0
except:
    print("Error occurred")
```

```python
try:
    x = int("abc")
except ValueError:
    print("Invalid number")
```

```python
try:
    x = int("abc")
    y = 10 / 0
except ValueError:
    print("Conversion error")
except ZeroDivisionError:
    print("Division error")
```

# finally Block
#### 👉 Hamesha run hota hai

```python
try:
    x = 5
except:
    print("Error")
finally:
    print("Done")
```

# else Block
#### 👉 Agar error na aaye to run hota hai

```python
try:
    x = 10 / 2
except:
    print("Error")
else:
    print("Success")
```

# Real-Time Example:

```python
try:
    num = int(input("Enter number: "))
    print(10 / num)
except ValueError:
    print("Invalid input")
except ZeroDivisionError:
    print("Cannot divide by zero")
```
