# typecasting

```python
import numpy as np
arr=np.array([1,2,3,4])
print(arr.dtype)
```

```python
arr=arr.astype(float)
print(arr)
print(arr.dtype)
```

```python
arr=np.array([2.3,3.3,4.3,4.5])
arr=arr.astype(int)
print(arr)
```
