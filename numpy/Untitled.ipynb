# saving and loadinig of numpy array

```python
import numpy as np

arr=np.array([1,2,3,4])
np.save("data.npy",arr)
np.load("data.npy")
```

```python
a=np.array([1,2,3,4])
b=np.array([5,6,7,8])
np.savez("multiple.npz",a=a,b=b)
data=np.load("multiple.npz")
print(data["a"])
print(data["b"] )     
```
