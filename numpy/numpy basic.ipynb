# dimensions

```python
#1d
a=[[[1,2,3,4]]]
arr=np.array(a)
print(arr)
print(arr.ndim)#ndim is use to tell us what is the dimension
```

```python
#2d
a=[[1,2,3],[4,5,6]]
arr=np.array(a)
print(arr)
print(arr.ndim)
```

```python
#3d
a=[[[1,2,3],[4,5,6],[7,8,9]]]
arr=np.array(a)
print(arr)
print(arr.ndim)
```

```python
#nd many dimensions no exist
a=[1,2,3,4]
arr=np.array(a,ndmin=5)
print(arr)
print(arr.ndim)
```
