# median

```python
import numpy as np
data=[22,22,55,12,199,12,19]
median=np.median(data)
print(median)
```

```python
import statistics as stat
data = [22,22,55,12,199,12,19]
median=stat.median(data)
print(median)
```

```python
import pandas as pd
data=[22,22,55,12,199,12,19]
s=pd.Series(data)
print(s.median())
```

```python
import seaborn as sns
import numpy as np
iris=sns.load_dataset("iris")
median=np.median(iris["sepal_length"])
print(median)
```

```python
import statistics as stat
number=[5, 8, 1, 7, 9, 3]
median=stat.median(number)
print(median)
```

```python
import statistics
numbers = input("Enter numbers separated by spaces: ")

num_list = [float(x) for x in numbers.split()]

median_value = statistics.median(num_list)
print("The median is:", median_value)
```
