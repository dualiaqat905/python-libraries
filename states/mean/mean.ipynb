# population mean

```python
population=[1,2,3,4,5,6,7]
pop=sum(population)/len(population)
print(pop)
```

# sample mean

```python
sample=[1,2,3,4]
pop=sum(sample)/len(sample)
print(pop)
```

# use numpy to find mean

```python
import numpy as np
sample=[15,24,13,12,11]
mean=np.mean(sample)
print(mean)
```

# use statistics to find mean

```python
import statistics as stat
me=[15,24,13,12,11]
mean=stat.mean(me)
print(mean)
```

```python
import pandas as pd
me=[15,24,13,12,11,2]
s=pd.Series(me)

print(s.mean())
```
