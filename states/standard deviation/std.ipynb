# standard deviation

```python
# sample std
import numpy as np 
data = [2,5,6,6,7,5,9,14,10,2,3,4]
sample=np.std(data,ddof=1)
print(sample)
```

```python
# population std
import numpy as np 
data = [2,5,6,6,7,5,9,14,10,2,3,4]
population=np.std(data,ddof=0)
print(population)
```

```python
import pandas as pd
data = [2,5,6,6,7,5,9,14,10,2,3,4]
series=pd.Series(data)
sample=series.std(ddof=1)
print(sample)
```

```python
population=series.std(ddof=0)
print(population)
```

```python
from scipy import stats

data = [2,5,6,6,7,5,9,14,10,2,3,4]

std = stats.tstd(data,ddof=0)
print(std)
```

```python
std = stats.tstd(data,ddof=1)
print(std)
```
