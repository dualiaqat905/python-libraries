# variance

```python
# sample variance
import numpy as np
data=[2,3,4,5,7,8,10,3,4,5]
sample=np.var(data,ddof=1)
print(sample)
```

```python
#population variance
import numpy as np
data=[2,3,4,5,7,8,10,3,4,5]
population=np.var(data,ddof=0)
print(population)
```

```python
import pandas as pd

data = [2,5,6,6,7,5,9,14,10,2,3,4]

series = pd.Series(data)
sample_variance = series.var()
print(sample_variance)

population_variance = series.var(ddof=0)
print(population_variance)
```
