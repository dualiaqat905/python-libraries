```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[20,10,40,50]
plt.plot(x,y,color="orange",linestyle="--",linewidth=3,marker="o",label="dua sales")
plt.xlabel("days")
plt.ylabel("sales")
plt.title("monthly saale")
plt.legend(loc="upper left")
plt.grid(color="grey",linestyle=":",linewidth=1)
plt.xlim(0,4)
plt.ylim(1,60)
plt.show()
```

```python
x=["monday","tuesday","wednesday","thursday","friday","saturday"]
y=[200,300,400,700,800,100]
plt.plot(x,y,color="orange",linestyle="--",linewidth=3,marker="o",label="daily  sales")
plt.xlabel("days")
plt.ylabel("sales")
plt.title("daily sale")
plt.legend(loc="upper left")
plt.grid(color="grey",linestyle=":",linewidth=1)
plt.xlim(0,6)
plt.ylim(0,1000)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[20,10,40,50]
plt.bar(x,y,color="orange",label="dua sales")
plt.xlabel("days")
plt.ylabel("sales")
plt.legend()
plt.show()
```

```python
region = ["north", "south","East","west"]
revenue= [3000,2000,1500,1000]
plt.pie(revenue,labels=region,autopct="%1.1f%%",colors=["yellow","blue","purple","grey"])
plt.title("revenue contribution")
plt.show()
```
